```python
# load of pre-trained model
import torch

from reasoning_from_scratch.ch02 import get_device
from reasoning_from_scratch.ch03 import (
    load_model_and_tokenizer,
)

device = get_device()

model, tokenizer = load_model_and_tokenizer(
    which_model="base",
    device=device,
    use_compile=False,
)
```

```python
# prep dataset
from llms_from_scratch.ch07 import download_and_load_file

file_path = "instruction-data.json"
url = "https://raw.githubusercontent.com/rasbt/LLMs-from-scratch/main/ch07/01_main-chapter-code/instruction-data.json"


data = download_and_load_file(file_path, url)
print("Number of entries:", len(data))
```

```python
from llms_from_scratch.ch07 import format_input

model_input = format_input(data[50])
pprint(model_input)
```

```python
train_portion = int(len(data) * 0.85)  # 1
test_portion = int(len(data) * 0.1)  # 2
val_portion = len(data) - train_portion - test_portion  # 3

train_data = data[:train_portion]
test_data = data[train_portion : train_portion + test_portion]
val_data = data[train_portion + test_portion :]

print("Training set length:", len(train_data))
print("Validation set length:", len(val_data))
print("Test set length:", len(test_data))
```

```python
from reasoning_from_scratch.ch03 import generate_text_stream_concat

input_text = format_input(val_data[8])
print("input_text:", input_text)


generate_text_stream_concat(
    model,
    tokenizer,
    input_text,
    device,
    20,
    verbose=False,
)
```
